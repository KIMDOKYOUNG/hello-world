```python
base_url = 'http://www.coupang.com'
```

```python
res = requests.get(base_url)
```

```python
res
```

```python
html = BeautifulSoup(res.content,'html.parser')
```

```python
html
```

```python
table = html.find('footer',{'id':'footer','class' :'footer_new'})
```

```python
table
```

```python
table2=table.find('div',{'class','footer-layer1'})('a')
```

```python
table2[1]['href'] #이미 a태그까지 뽑은 상태이므로 바로 href를 뽑는다
```

```python
len(table2)
```

```python
link =[]
for i in range(len(table2)):
    ob_link=table2[i]['href']
    link.append(ob_link)
```

```python
link
```

```python
link[2]
```

```python
res2 = requests.get(link[2])
```

```python
html2 = BeautifulSoup(res2.content,'html.parser')
```

```python
table = html2.find('ul',{'id':'csNoticeMenuList','class':'cs-notice__menu-list'})
```

```python
table
```

```python
gong_url = link[2]+table.find_all('a',{'class':'cs-notice__menu-link'})[1]['href']
```

```python
res = requests.get(gong_url)
html = BeautifulSoup(res.content,'html.parser')
```
