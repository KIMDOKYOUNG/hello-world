#### https://plotnine.readthedocs.io/en/stable/api.html
#### http://r-statistics.co/Complete-Ggplot2-Tutorial-Part1-With-R-Code.html
- data : 변수 매핑을 설명하는 aes 매핑의 데이터 셋
- geom : 점, 선, 막대 등 plot에서 나타내고 싶은 정보
- stats : 다양한 방법으로 데이터를 요약 집계하여 표현
- scale : aes 공간에 데이터 값을 매핑
- coord : 데이터 좌표가 그래픽 면에 매핑되는 방법
- facet : 데이터를 원하는 단위로 분해하는 방법

## 1. basic plot
- ggplot(data, aes(x, y, size, color)) + geom

### 1-(1) point plot

```python
# x,y 축 모두 연속일 때
ggplot(log, aes(x = 'TOTAL_DATE_PV', y = 'PAGEVIEWS')) + geom_point()
```

```python
# x,y 축 연속형, 범주형 정보 하나 추가하고 싶을때
(
    ggplot(log)
        + aes(x = 'TOTAL_DATE_PV', y = 'PAGEVIEWS') 
        + geom_point(aes(size = 'AGE'))
)
```

```python
(
    ggplot(log) 
        + aes(x = 'TOTAL_DATE_PV', y = 'PAGEVIEWS') 
        + geom_point(aes(color = 'AGE'))
)
```

```python
# x,y 축 연속형, 범주형 정보 두개 추가하고 싶을때
(
    ggplot(log) 
        + aes(x = 'TOTAL_DATE_PV', y = 'PAGEVIEWS') 
        + geom_point()
        + aes(size ='AGE', color = 'GENDER')
)
```

```python
(
    ggplot(log) 
        + aes(x = 'TOTAL_DATE_PV', y = 'PAGEVIEWS') 
        + geom_point(size = 1)
        + aes(color = 'GENDER')
)
```

```python
# 추세(trend)를 보고 싶을때
(
    ggplot(log) 
        + aes(x = 'TOTAL_DATE_PV', y = 'PAGEVIEWS') 
        + geom_point(size = 1)
        + aes(color = 'GENDER')
        + stat_smooth()
        + ylim(0,100000)
)
```

```python
# 범주별로 각각 그리고 싶을때
(
    ggplot(log) 
        + aes(x = 'TOTAL_DATE_PV', y = 'PAGEVIEWS') 
        + geom_point(size = 1)
        + aes(color = 'GENDER')
        + stat_smooth()
        + facet_wrap('AGE', ncol = 3)
        + ylim(0,100000)
)
```

## 시각화 활용

```python
DF = log >> mutate(TIME_AM = X.TIME_0to5 + X.TIME_6to11, TIME_PM = X.TIME_12to17 + X.TIME_18to23) >> select(X.AGE, X.TIME_AM, X.TIME_PM, contains('WEEKDAY')) >> drop(-1)
```

```python
DF
```

```python
DF_NEW = DF >> select(columns_between(0,3),'WEEKDAY_금')

(
    ggplot(DF_NEW) 
        + aes(x = 'WEEKDAY_금', y = 'TIME_AM')
        + geom_point(size = 1)
        + aes(color = 'AGE')
        + stat_smooth()
        + facet_grid('AGE~')
        + theme(text=element_text(family=font_name),
                figure_size = (15,10),
                axis_text_x = element_text(angle=45))
)
```

```python
def plotss(DF, WEEKDAY, TIME):
    DF_NEW = DF >> select(columns_between(0,3),WEEKDAY)

    plot =     (
                    ggplot(DF_NEW) 
                        + aes(x = WEEKDAY, y = TIME)
                        + geom_point(size = 1)
                        + aes(color = 'AGE')
                        + stat_smooth()
                        + facet_grid('AGE~')
                        + theme(text=element_text(family=font_name),
                                figure_size = (15,10),
                                axis_text_x = element_text(angle=45))
                )
    return plot
```

```python
weekday_list = DF.columns[3:][[3,6,2,1,0,5,4]]
```

```python
plotss(DF, weekday_list[1], 'TIME_AM')
```
