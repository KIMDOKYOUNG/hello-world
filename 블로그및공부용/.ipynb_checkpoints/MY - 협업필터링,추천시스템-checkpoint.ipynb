# 베이스라인 모형

u = user / i = item
exp(Rui) = mean(전체평점) + Bu(동일한 사용자에 의한 평점 조정값) + Bi(동일한 상품에 대한 평점 조정값)
즉, 베이스라인의 모형은 오차함수를 최소화 하도록 구해짐 

베이스라인 모형은 오차함수를 최소화 하기 위해서 SGD / ALS 라는 최적화 method를 제공함

SGD (Stochastic Gradient Descent)의 인수
    reg: 정규화 가중치. 디폴트는 0.02.
    learning_rate: 최적화 스텝 사이즈. 디폴트는 0.005.
    n_epochs: 최적화 반복 횟수. 디폴트는 20.
    
ALS (Alternating Least Squares)의 인수
    reg_i: 상품에 대한 정규화 가중치. 디폴트는 10.
    reg_u: 사용자에 대한 정규화 가중치. 디폴트는 15.
    n_epochs: 최적화 반복 횟수. 디폴트는 10.

보통의 surprise 패키지를 사용할 때의 순서
1. 데이터세트의 split, folds 메소드를 사용하여 K-Fold 트레이닝 데이터셋과 테스트 데이터셋을 만든다.
2. 모형 알고리즘 객체를 생성한다.
3. 모형 알고리즘 객체의 train 메서드와 트레이닝 데이터셋으로 모수를 추정한 후, test 메서드로 테스트 데이터셋에 대한 예측을 실시한다.
4. accuracy 서브패키지의 성능평가 함수를 사용하여 예측 성능을 계산한다.

# collaborative filtering

#### 데이터 구성 : 사용자가 입력한 선호도를 사용하여 사용자-item의 선호도행렬을 만든다 - 평점행렬이 가지는 특정한 패턴을 찾아서 이를 평점 예측에 사용
#### 상품 기준으로 평점의 유사성을살피는 neighborhood / 행렬의 수치적 특징을 이용하는 latent factor 모형이 존재
#### 유사도 계산 : 1단계에서 만들어진 행렬을 사용하여 사용자들간의 유사도 계산(euclidian,cosine)
#### 예측 값 계산 및 추천 목록 생성 : 사용자들 간의 유사도를 바탕으로 모든 항목에 대해 예측값을 계산하고 높은 예측값을 갖는 상위 n개의 추천 목록을 생성한다.


#### neighborhood 모형
-특정 사용자의 평점을 예측하기 위해 사용되는 것이 아니라 해당 사용자와 유사한 사용자에 대한 가중치로 추천.

-해당 사용자와 유사한 사용자의 행 벡터를 찾아서 이를 기반으로 비어있는 데이터를 예측하는 user-based cf라고 함(행을 기준으로 보자!)

-특정한 상품에 대해 사용자가 준 점수 즉, 평점 행렬의 상품 열 벡터의 유사성을 찾고 특정 상품과 유사한 평점 정보를 가지는 상품들로 해당 상품의 빈 데이터 예측하는 것은 item-based cf라고 함 (열을 기준으로 보자!)

#### 유사도 계산
평균제곱차이 유사도 (Mean Squared Difference Similarity)
코사인 유사도 (Cosine Similarity)
피어슨 유사도 (Pearson Similarity)
피어슨-베이스라인 유사도 (Pearson-Baseline Similarity)

#### 평균 제곱차이 유사도
msd = 유클리드 공간에서의 거리제곱에 비례하는 값
유사도 = 1/msd 로써 msd가 클수록 유사도는 떨어진다.

#### 코사인 유사도
코사인 유사도는 세타 즉, 각도에 의해서 유사도를 판단한다. 세타가0이면 유사도는1이되고 세타가90도이면 유사도는 0이다

#### 피어슨 유사도
두벡터의 상관계수로 정의 - 상관계수가 높을수록 유사도가 높음

#### 피어슨 베이스라인 유사도
피어슨 유사도처럼 상관계수를 구하는 것이지만 각 벡터의 기댓값을 평군이 아닌 베이스라인 모형에서 예측한 값을 사용
피어슨-베이스라인 유사도는 벡터의 차원 즉, 두 사용자나 상품에 공통적으로 있는 평점 원소의 갯수를 이용하여 정규화를 하는 shrinkage를 추가하여 사용한다.

#### KNN가중치 예측 방법
일단 유사도가 구해지면 평점을 예측하고자 하는 사용자(또는 상품)와 유사도가 큰  k 개의 사용자(또는 상품) 벡터를 사용하여 가중 평균을 구해서 가중치를 예측한다. 이러한 방법을 KNN(K Nearest Neighbors) 기반 예측 방법이라고 한다.
- 가중치 예측 알고리즘 클래스
    KNNBasic = 평점들을 단순히 가중 평균
    KNNWithMeans = 평점들을 평균값 기준으로 가중 평균
    KNNBaseline = 평점들을 베이스 라인 모형의 값 기준으로 가중 평균

# 데이터 학습 및 예측 (baselineonly)

```python
bsl_options = {'method': 'als',
              'n_epochs':5,
              'reg_u': 12,
              'reg_i':5}
algo = BaselineOnly(bsl_options = bsl_options)#parameter처리
cross_validate(algo,data,measures=['RMSE'],cv=3,verbose=True)
```

rmse정확도 훈련 셋과 검증셋을 샘플링하기 위해서 train_test_split()를 사용하여 rmse정확도를 척도로 사용
fit() method를 통해서 훈련데이터셋의 알고리즘을 훈련시키고 test를 통해 검증셋으로부터 생성된 예측을 반환

```python
trainset, testset = train_test_split(data,test_size=0.25)

algo = BaselineOnly(bsl_options = bsl_options)
predictions = algo.fit(trainset).test(testset)

accuracy.rmse(predictions)
```

predictions의 r_ui는 실제 평점 / est는 estimate의 약자로 예측값을 나타낸다

```python
predictions
#나온 결과와 추천알고리즘 저장
#dump.dump('./dump_file',predictions,algo)
#predictions,algo = dump.load('./dump_file')
```

```python
trainset =  algo.trainset
```

```python
print(algo.__class__.__name__)
```

```python
def est_ui(userid):
    try:
        return len(trainset.ur[trainset.to_inner_uid(userid)])
    except:
        return 0
    
def est_iu(itemid):
    try:
        return len(trainset.ur[trainset.to_inner_iid(itemid)])#학습모델을 통해서 iid가 필요하므로
    except:
        return 0
    
df = pd.DataFrame(predictions,columns = ['uid','iid','rui','est','details'])
df['iu'] = df['uid'].apply(est_ui)
df['iu'] = df['iid'].apply(est_iu)
df['err'] = abs(df['rui']-df['est'])

```

### popularity, high rated based 추천 시스템
#### 가장 쉽게 인기도, 높은 평점을 갖는 item을 추천가능
#### 모두에게 동일한 item이 추천

```python
#데이터를 튜플형식으로 나타냄
ratings={
     'Dave':{'달콤한인생':5,'범죄도시':3,'샤인':3},
     'David':{'달콤한인생':5,'범죄도시':1,'샤인':4},
     'Alex':{'달콤한인생':0,'범죄도시':4,'샤인':5},
     'Andy':{'달콤한인생':2,'범죄도시':1,'샤인':5}
}
ratings['Dave']#['달콤한인생']
```

#### 즉 해당 추천 시스템에서는 평점이 높은순서대로 item을 추천해주면 됨

```python

#영화들의 평점 합을 구하기
movie_dict = dict()
for rating in ratings:
    for movie in ratings[rating].keys():
        if movie not in movie_dict:
            movie_dict[movie] = ratings[rating][movie]
        else:
            movie_dict[movie] = (movie_dict[movie] + ratings[rating][movie]) 

#영화별로 평균 내기
for i in movie_dict.keys():
    movie_dict[i] = movie_dict[i]/4
    
import operator
#영화별 아이템화 한 후 평점 높은 순 2개 추천해주기
#튜플 형식을 내림차순 정렬한 후 리스트화
sorted_x = sorted(movie_dict.items(),key=operator.itemgetter(1),reverse=True)

sorted_x[0:2]
```

# 협업필터링 유사도 기반 추천 시스템

```python
ratings={
     'Dave':{'달콤한인생':5,'범죄도시':3,'샤인':3},
     'David':{'달콤한인생':2,'범죄도시':1,'샤인':4},
     'Alex':{'범죄도시':4,'샤인':5},
     'Andy':{'달콤한인생':2,'범죄도시':1,'샤인':5}
}
```

```python
#딕셔너리형식은 .get method를 통해서 value산출 가능
ratings['Dave'].get('샤인')
```

유사도구하기(피타고라스)

```python
import math
```

```python
def sim(i,j):
    return math.sqrt(pow(i,2)+pow(j,2))#pow(a,b) = a**b
```

```python
var1 = ratings['Alex'].get('범죄도시') -ratings['Andy'].get('범죄도시')
```

```python
var2 =  ratings['Alex'].get('샤인') -ratings['Andy'].get('샤인')
```

```python
sim(var1,var2)
```

alex가 평가한 범죄도시, 샤인을  평가한 다른 사용자와의 거리 구하기(유사도 구하기)

```python
for i in ratings:
    if i == 'Alex':
        pass
    else:
        var1 = ratings['Alex'].get('범죄도시') -ratings[i].get('범죄도시')
        var2 = ratings['Alex'].get('샤인') -ratings[i].get('샤인')
        print(i + ' : ' + str(sim(var1,var2)))
```

유사도가 0-1사이에 위치하고 1과 가까울수록 유사도가 높아지도록 변경

```python
for i in ratings:
    if i == 'Alex':
        pass
    else:
        var1 = ratings['Alex'].get('범죄도시') -ratings[i].get('범죄도시')
        var2 = ratings['Alex'].get('샤인') -ratings[i].get('샤인')
        print(i + ' : ' + str(1/(1+sim(var1,var2))))
```

# 유사도 계산 방식

1. 평균제곱차이 유사도(msd)  
2. 코사인 유사도  
3. 피어슨 유사도

#### 1. 평균제곱차이 유사도  
* user-based collaborative filter  
사용자 u와 v간의 msd = 사용자 u,v가 평가한 상품들의 평점간의 차의 제곱 / 사용자 u,v모두 평가한 **상품의 수**    
* item-based collaborative filter  
분모를 상품i와 상품j모두를 평가한 **사용자**의 수  
따라서 유사도를 구하기 위해서는 1에 가까울수록 높게 만들기 위해서 역수 취급으로 하여 유사도를 구한다

```python
def sim_msd(data,name1,name2):
    sum = 0
    count = 0
    for movie in data[name1]:
        if movie in data[name2]: #name1유저와 name2유저가 같은 영화를 봤는지 확인하기 위해서
            sum += pow(data[name1][movie] - data[name2][movie],2)
            count += 1
            
    return sum/count
            
```

```python
1/(1+sim_msd(ratings,'Dave','Alex')) #데이브와 알렉스의 유사도
```

```python
1/(1+sim_msd(ratings,'Dave','Andy'))#데이브와 앤디의 유사도
```

#### 2. 코사인유사도
* 코사인 유사도는 두 특성 벡터간의 유사 정도를 코사인 값으로 표현한 것 -> w2v에서 주로 사용  
* 코사인 유사도는 [-1,1]의 범위값을 가지며 -1은 완전히 반대되는 경우 0은 독립 1은 완전이 같은 경우  
* cos = x * y / abs(x) * abs(y)  
* 코사인유사도도 마찬가지로 같은 영화를 본 것에 대해서 계산을 해야한다
* cosine_sim(u,v) = sum((u의평점) * (v의평점)) / sqrt(sum(u의평점^2)) * sqrt(sum(v의평점^2))  
* cosine_sim(i,j) = 상품간의 유사도도 마찬가지  
* 스칼라곱 - 두벡터로 스칼라를 계산하는 연산 a벡터 * b 벡터 = sum(ai * bi)
* 따라서 w2v에서 변수간의 코사인 유사도를 도출할 때 위의 식으로 도출하는 것이다 열을 기준으로  
* a열과 b열간의 유사도를 구하는 것이기 때문에 코사인 유사도를 구하려면 열을 기준으로 구하자 생각해보니 행도 가능할듯?  

```python
ratings
```

```python
def sim_cosine(data,name1,name2):
    cos1 = 0
    cos2 = 0
    cos1_cos2 = 0
    for i in data[name1]:
        if i in data[name2]:
            cos1 += pow(data[name1][i],2)
            cos2 += pow(data[name2][i],2)
            cos1_cos2 += data[name1][i]*data[name2][i]
    return cos1_cos2/(math.sqrt(cos1)*math.sqrt(cos2))
```

```python
#dave와 가장 유사한 사용자는?
sim_cosine(ratings,'Dave','Alex')
```

```python
sim_cosine(ratings,'Dave','David')
```

```python
sim_cosine(ratings,'Dave','Andy')
```

#### 피어슨 유사도
* 피어슨 유사도는 두 벡터의 상관계수를 의미한다  
* 특정인물의 점수기준이 극단적으로 너무 낮거나 높을 경우 유사도에 영향을 크게 주기 때문에 이를 막기 위해 상관계수를 사용  
* user based pearson sim =  
                           sum((u의평점-u의평균평점) * (v의평점-v의평균평점)) /     sqrt(sum((u의평점-u의평균평점)^2)) * sqrt(sum((v의평점-v의평균평점)^2))  
* item기반도 마찬가지

```python
def sim_pearson(data, name1, name2):
    avg_name1 = 0
    avg_name2 = 0
    count = 0
    for movies in data[name1]:
        if movies in data[name2]: #같은 영화를 봤다면
            avg_name1 = data[name1][movies]
            avg_name2 = data[name2][movies]
            count += 1
    
    avg_name1 = avg_name1 / count
    avg_name2 = avg_name2 / count
    
    sum_name1 = 0
    sum_name2 = 0
    sum_name1_name2 = 0
    count = 0
    for movies in data[name1]:
        if movies in data[name2]: #같은 영화를 봤다면
            sum_name1 += pow(data[name1][movies] - avg_name1, 2)
            sum_name2 += pow(data[name2][movies] - avg_name2, 2)
            sum_name1_name2 += (data[name1][movies] - avg_name1) * (data[name2][movies] - avg_name2)
    
    return sum_name1_name2 / (math.sqrt(sum_name1)*math.sqrt(sum_name2))
```

```python
sim_pearson(ratings,'Dave','Alex')
```

```python
def top_match(data,name,index=3,sim_function=sim_pearson):
    li = []
    for i in data:
        if i != name:
            li.append((sim_function(data,name,i),i))#sim_function()의 피어슨 상관계수 추출
    li.sort()#정렬
    li.reverse()#내림차순
    return li[:index]
```

```python
top_match(ratings,'Dave',3)
```

```python
top_match(ratings,'Dave',3,sim_function=sim_cosine)
```
