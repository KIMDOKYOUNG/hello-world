# RNN의 기본 구조와 Keras를 사용한 RNN구현

신경망을 사용하여 문장이나 시계열 데이터와 같은 순서열을 예측하는 문제를 푸는 경우, 예측하고자 하는 값이 더 오랜 과거의 데이터에 의존하게 하려면 sequence를 나타내는 벡터의 크기를 증가시켜야 한다.
여기서 RNN은 뉴런의 상태(St)를 저장하고 이를 다음 스텝에서의 입력으로 사용함으로써 긴 순서열에 대해서도 예측을 할 수 있는 신경망 구조이다. 

## RNN의 기본구조

일반적인 feedforward신경망 구조는 아래와 같이 출력벡터 y가 입력 x벡터와 신경망 가중치 행렬 u의 곱에 허ㅏㄹ성화함수를 적용한 결과로 나타난다.
-sigma(Ux)  : sigma는 활성화함수를 뜻한다.

하나의 은닉층을 가지는 MLP의 경우에는 아래와 같이 표현가능
-h = sigma(Ux)   :  h는 은닉층 벡터, U는 입력으로부터 은닉층까지 가중치 행렬
-o = sigma(Vh)   :  V는 은닉층으로부터 출력까지의 가중치 행렬 , o는 출력 벡터

RNN에서는 출력벡터 o이외에도 상태 벡터 s를 출력한다. 상태 벡터는 일종의 은닉층 벡터와 비슷하지만 입력x뿐 아니라 바로 전단계의 상태 벡터 값에도 의존한다. 즉, 출력벡터는 상태벡터의 값에 의존한다.
- St = sigma(Uxt + WSt-1)  :  W는 상태벡터에 영향을 주는 가중치 벡터 , St-1은 현재 입력벡터 xt에 영향을 주는 전단계의 상태벡터
- ot = sigma(VSt)

RMM은 시간 스텝에 따라서 연결해서 펼쳐놓으면 무한개의 은닉층을 가진 MLP와 유사한 효과가 있다.

## 순서열 예측

RNN의 순서열을 예측하는 방법에는 3가지가 있는데 many to one / one to many / many to many가 있다.

만약, 원하는 결과가 출력 순서열 o1~ot가 target순서열 y1~yt과 같아지는 것이라면 입력 순서열 길이와 출력 순서열 길이가 같은 특수한 경우의 many to many 예측 문제가 된다.

RNN은 시간에 따라 펼쳐놓으면 구조가 MLP와 유사하기 때문에 Back-propagation방법으 gradient를 계산할 수 있다.그렇지만 실제로 여러개의 은닉층이 있는 것이 아니라 시간 차원에서 존재하기 때문에 BPTT방법이라고 함

## RNN의 문제

적절한 정보와 그 정보가 필요한 곳과의 차이가 적을 경우 RNN은 과거 정보를 이용하여 잘 학습할 수 있게 됩니다. 예를 들어 단어의 나열이 input라고 생각했을 때 문맥을 이용할 때에 중요한 단어와 단어 사이의 gap이 적을 경우 위의 설명처럼 rnn이 잘 동작할테지만 만약 그 gap이 넓을 경우에는 rnn은 잘 동작하지 못한다. = 장기의존성 문제
-> 이러한 단점을 보완한 것이 LSTM

## LSTM
https://blog.naver.com/kmkim1222/221069000196

RNN의 장기의존성을 보완하여 학습할 수있는 RNN의 한 종류이며 오랫동안 정보를 기억하는 기능을 가지고 있는 것이다.

LSTM은 기존 vanilla RNN을 개선한 모델로써 데이터의 long-term dependency를 학습하는데 효과적인 모델입니다. 또한 내부적으로 구성요소간에 additive, elementwise interaction을 통해 backpropagation 할 때, gradient explosion, vanishing 문제를 기존 RNN에 비해 매우 줄인 것이 중요한 개선점이다.

# Keras를 통한 LSTM구현

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

plt.style.use('ggplot') #ggplot스타일 사용
```

```python
df = pd.read_csv('LSTM_study.csv',engine='python',skiprows=6, skipfooter=9)
```

```python
df.head()
```

#### 날짜데이터 전처리

```python
from pandas.tseries.offsets import MonthEnd
```

```python
df['Adjustments'] = pd.to_datetime(df['Adjustments']) + MonthEnd(1) 
#년도와 월밖에 나와 있지 않은 데이터를 해당 월의 가장 마지막 일로 데이터를 만들어 준다.
```

```python
df = df.set_index('Adjustments')
```

```python
df.plot(figsize = (12,10))
```

### 훈련/테스트 데이터 분할하기
2011/1/1을 기준으로 training set과 test set으로 나눠서 구분한다.
그리고 사용할 변수는 unadjustments

```python
split_date = pd.Timestamp('01-01-2011')
#2011/1/1 까지의 데이터를 train데이터  / 그 이후 데이터를 test데이터로 구분

train = df.loc[:split_date,['Unadjusted']]
test = df.loc[split_date:,['Unadjusted']]
```

```python
plt.figure(figsize=(10,8))
train['Unadjusted'].plot()
test['Unadjusted'].plot()
plt.legend(['train','test'])
```

### 변수 scaling
min-max scaling을 통해서 변수스케일링

```python
from sklearn.preprocessing import MinMaxScaler

```

```python
sc = MinMaxScaler()

#sc.transform - 테스트데이터에 실행 / sc.fit_transform - 훈련데이터에 실행
train_sc = sc.fit_transform(train)
test_sc = sc.transform(test)
```

```python
train_df = pd.DataFrame(train_sc,columns = ['Scaled'],index = train.index)
test_df = pd.DataFrame(test_sc,columns = ['Scaled'],index = test.index)
```

```python
for i in range(1,13):
    train_df['shift_{}'.format(i)] = train_df['Scaled'].shift(i)
    test_df['shift_{}'.format(i)] = test_df['Scaled'].shift(i) 
```

```python
x_train = train_df.dropna().drop('Scaled',axis=1)
y_train = train_df.dropna()['Scaled']

x_test = test_df.dropna().drop('Scaled',axis=1)
y_test = test_df.dropna()['Scaled']
```

```python
x_train.head()
```

### 딥러닝 모델에 데이터를 넣기 위해서 pandas구조의 데이터를 ndarray로 변환시켜준다.
->ndarray로 변환시켜주는 명령어는 dataframe.values

```python
x_train = x_train.values
y_train = y_train.values
x_test = x_test.values
y_test = y_test.values
```

### 최종 트레이닝셋과 테스트셋의 x만들기

keras에서 RNN계열의 모델을 트레이닝할 때 요구되는 데이터 형식이 있음. 그것은 바로 3차원 데이터여야 ㅎ하며 각각의 차원은(size,timestep,feature)을 순서대로 나타내주어야 한다.
따라서 이 형태로 데이터를 reshape해줘야하며 다중퍼셉트론 모델에서는 size와 feature만 있기 때문에 2차원이지만, RNN에서는 "시간"이라는 개념이 있기 때문에 차원이 한 차원 늘어나게 된것입니다. 

그래서 훈련데이터와 테스트데이터의 X를 RNN학습에 맞는 데이터형식으로 또 변환해줘야한다.

```python
X_train_t = x_train.reshape(x_train.shape[0], 12, 1) 
X_test_t = x_test.reshape(x_test.shape[0], 12, 1)

```

# LSTM 모델 만들기 -keras 활용

아래와 같이 keras를 통해서 LSTM모델을 만들 수 있습니다.
input_shape = (timestep,feature)으로 만들어줍니다. size는 모델 설계시에는 중요x
모델에는 timestamp, feature만 알려주면 된다.

-예측하고자 하는 target의 갯수가 1이므로 마지막 출력층 dense에 1을 추가해준다. 
-연속형 변수를 예측하는 것이기 때문에 loss function 은 MSE를 척도로써 사용
-optimizing는 adam을 주로사용

```python
from keras.layers import LSTM 
from keras.models import Sequential 
from keras.layers import Dense 
import keras.backend as K 
from keras.callbacks import EarlyStopping 
```

```python
K.clear_session()
model = Sequential() #sequential 모델
model.add(LSTM(20, input_shape = (12,1))) #lstm(timestamp, feature)
model.add(Dense(1)) # output = 1 
model.compile(loss='mean_squared_error', optimizer='adam')

model.summary()
```

### 데이터를 이용하여 model fitting

모델을 Fitting한다는 것은 Training data set으로 optimization 과정을 통해 모델의 weight를 찾는 것입니다. 
early stopping 객체를 이용해 epoch마다 early stopping을 체크합니다. 
아래와 같은 결과를 내면서 트레이닝이 잘 되는 것을 확인할 수 있습니다.

```python
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)

model.fit(X_train_t, y_train, epochs=100,
          batch_size=30, verbose=1, callbacks=[early_stop])
```

### 훈련데이터로 학습한 모델을 이용하여 test set 검증

```python
y_pred = model.predict(X_test_t)
```

```python
plt.plot(test_df.dropna().index,y_pred)
plt.plot(test_df.dropna().index,y_test)
```

# RNN을 활용한 시계열 예측문제

아래의 그림과 같은 사인파형 시계열을 입력으로 다음 스텝의 출력을 예측하는 단순 시계열 예측 문제

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import FinanceDataReader as fdr
```

```python
plt.style.use('ggplot') #ggplot스타일 사용
```

```python
s = np.sin(2 * np.pi * 0.125 * np.arange(20)) #20개의 값을 출력 -  sin 함수 이용
```

```python
plt.figure(figsize=(10,8))
plt.plot(s,'ro-')
plt.xlim(-0.5,20.5)
plt.ylim(-1.1,1.1)
plt.show()
```

이제 위의 시계열 자료에서 입력 순서열과 target값을 만든다. 
입력 순서열 - 3스텝크기의 순서열 사용
target - 그 다음 시간 스텝의 값을 사용

3개의 순서열을 입력한 다음 마지막 출력값이 target과 일치하게 만드는 sequence-to-value (many-to-one)문제를 풀어보자

**Keras 에서 RNN 을 사용하려면 입력 데이터는 (nb_samples, timesteps, input_dim) 크기를 가지는 ndim=3인 3차원 텐서(tensor) 형태이어야 한다.

nb_samples: 자료의 수
timesteps: 순서열의 길이
input_dim: x 벡터의 크기

```python
#지금 해보려는 문제는 단일 시계열이므로 input_dim = 1이고 timesteps = 3이며 자료의 수는 18개이다.
#다음코드와 같이 원래의 시계열 벡터를 Toeplitz 행렬 형태로 변환하여 3차원 텐서를 만든다.
#np.fliplr - flipflop라고 생각하면 된다. 행렬을 완전히 뒤집어 주는 것
#덧붙여서 말하면 list요소를 가지고 있는 array에서 각 list원소를 거꾸로 만들어 주는 것
from scipy.linalg import toeplitz
S = np.fliplr(toeplitz(np.r_[s[-1], np.zeros(s.shape[0] - 2)], s[::-1]))
```

```python
np.r_[s[-1], np.zeros(s.shape[0] - 2)] #array형태로 append해주는 명령어
```

```python
S[:5,:3] #배열안의 5개의 요소를 추출하고 각 요소별 3번째 요소까지 추출하라는 것
```
