```python
base_url = 'http://www.weather.go.kr/weather/climate/past_tendays.jsp?stn=108&yy=2018&mm=4&obs=1&x=19&y=8'
```

```python
browser = Chrome()
browser.maximize_window()
browser.get(base_url)
```

```python
df4 = pd.DataFrame(columns=['월_일','평균기온1','평균기온2','최고기온1','최고기온2','최저기온1','최저기온2','일교차','최저초상온도','상대습도1','상대습도2','상대습도3','운량','일사','일조1','일조2','평균풍속','최대풍향','최대풍속','최대순간풍향','최대순간풍속','증발량소','증발량대','강수량1','강수량2','강수계속시간','신적설1','신적설2','적설1','적설2','지면온도','철관지중온도1','철관지중온도2','철관지중온도3','철관지중온도4','철관지중온도5'])
for mon in range(1,6):
    browser.find_elements_by_xpath('//*[@class= "observation_form"]/select[3]')[0].click()
    time.sleep(1.5)
    browser.find_elements_by_xpath('//*[@class= "observation_form"]/select[3]/option['+ str(mon) + ']')[0].click() #option에서 반복문
    time.sleep(1.5)
    
    for si in range(1,4):
        browser.find_elements_by_xpath('//*[@class= "observation_form"]/select[4]')[0].click()
        time.sleep(1.5)
        browser.find_elements_by_xpath('//*[@class= "observation_form"]/select[4]/option['+ str(si) + ']')[0].click() #option에서 반복문
        time.sleep(1.5)
        browser.find_elements_by_xpath('//*[@class= "observation_form"]/input[4]')[0].click()
        time.sleep(1.5)
        
        html0 = browser.page_source
        html1 = BeautifulSoup(html0,'html.parser')
        
        table = html1.find('table',{'class':'table_develop'})
        table2 = table.find_all('th',{'class':'top_line'})
        table3 = table2[1:]
        
        col = []
        for i in range(len(table3)):
            x= table3[i].get_text()
            col.append(x)
        
        col = col[:-2]
        col2 = ['평균기온1','평균기온2','최고기온1','최고기온2','최저기온1','최저기온2','일교차','최저초상온도','상대습도1','상대습도2','상대습도3','운량','일사','일조1','일조2','평균풍속','최대풍향','최대풍속','최대순간풍향','최대순간풍속','증발량소','증발량대','강수량1','강수량2','강수계속시간','신적설1','신적설2','적설1','적설2','지면온도','철관지중온도1','철관지중온도2','철관지중온도3','철관지중온도4','철관지중온도5']
        
        tbody = table.find('tbody')
        tr = tbody.find_all('tr')
        
        td2 = []
        for j in range(len(tr)):
            td3 = []
            for i in range(len(tr[0].find_all('td'))):
                if j==15 or j==17 or j==19:
                    td0 = tr[j].find_all('td')[i].text
                    td1 = ''.join(re.findall('[가-힣0-9.:-]',td0))
                    if len(td1)==7:
                        td11 = td1[0:3]
                    elif len(td1) ==5:
                        td11 = td1[0]
                    else:
                        td11 = td1[0:4] 
                    td3.append(td11)
                else:
                    td0 = tr[j].find_all('td')[i].text
                    td1 = ''.join(re.findall('[가-힣0-9.:-]',td0))
                    td3.append(td1)
            td2.append(td3[:-2])
        
        tem = list(map(list,zip(*td2)))
        df1 = pd.DataFrame(tem,columns=col2)
        df2 = pd.DataFrame(col,columns=['월_일'])
        
        df3 = pd.concat([df2, df1], axis=1)
        
        df4 = df4.append(df3)
df4.index = range(len(df4))
df4['연도'] = '2019'
```
